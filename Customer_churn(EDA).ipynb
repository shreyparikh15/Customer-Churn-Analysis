```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Imported libraries

# Reading the dataset

```python
df = pd.read_csv('Customer Churn.csv')
df
```

# Data cleaning

```python
df.shape
```

```python
df.info()
```

## Changing the data type of 'TotalCharges' column from 'object' to 'float'

### Replacing blanks with 0 as tenure is 0 and no total charges are recorded 

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", '0')
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

```python
df.info()
```

**Data type of 'TotalCharges' changed from object to float**

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

## Converted 0 and 1 values of senior citizen to 'yes/no' to make it easier to understand

```python
df['SeniorCitizen'] = df['SeniorCitizen'].map({'Yes': 1, 'No': 0})

```

```python
df.tail()
```

# Data Analysis

```python
ax = sns.countplot(x = 'Churn', data=df)        # 'x' means axis
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb = df.groupby('Churn').agg({'Churn':'count'})
gb
```

```python
ex = [0, 0.1]
plt.pie(gb['Churn'], labels=['Yes', 'No'], autopct = "%1.3f%%", explode=ex, shadow=True)
plt.title("Percentage of Churned Customeres", fontsize = 15)
plt.show()
```

***from the given pie chart we can conclude that 26.54% of our customers have churned out. 
#not let's explore the reason behind it***

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x = 'gender',data=df, hue='Churn')

# Loop through each container to label all bars
for container in ax.containers:
    ax.bar_label(container)
    
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,5))
ax = sns.countplot(x='SeniorCitizen', data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
ax = sns.countplot(x='Contract', data=df, hue='Churn')

for container in ax.containers:
    ax.bar_label(container)

plt.title("Count of Customers by Contract")
plt.xticks(rotation=30)
plt.show()
```

***Conclusion - People who have month to month contract are likely to churn then from those who have 1 or 2 years or contract*** 

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Plot
fig, ax = plt.subplots(figsize=(4, 4))  # Adjust figsize for better visualization

# Plot the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])  # Customize colors if desired

# Add percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor = (0.9,0.9))  # Customize legend location

plt.show()
```

***Comparative a greater pecentage of people in senior citizen category have churned***

```python
# Tennure Histogram using MATPLOTLIB
l = [10,20,30,40,50,60,70,80]
plt.hist(df['tenure'], bins=l, edgecolor='black', color='orange')

plt.title('Customer Tenure Period')
plt.xlabel('Tenure')
plt.ylabel('Value')
plt.show()
```

```python
# Tennure Histogram using SEABORN
plt.figure(figsize=(9,5))
sns.histplot(x='tenure', data=df, bins=72, hue='Churn')
plt.show()
```

***Conclusion - People who have used our services for a long time have stayed and people who have used our sevices 
#1 or 2 months  have churned***

```python
ax = sns.countplot(x='PaymentMethod', data=df, hue='Churn')

for container in ax.containers:
    ax.bar_label(container)

plt.xticks(rotation=30)
plt.title('"Churned Customers by Payment Method"')
plt.show()
```

***Conclusion - Customer is likely to churn when he is using electronic check as a payment method***

```python
# List of columns you want to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 
           'StreamingTV', 'StreamingMovies']

# Set the figure size based on the number of subplots
plt.figure(figsize=(16, 12))

# Loop through each column and create a subplot for each
for i, column in enumerate(columns, 1):
    plt.subplot(3, 3, i)  # Create 3x3 grid of subplots
    sns.countplot(x=column, data=df, hue='Churn')
    plt.title(f'Countplot of {column}')
    plt.tight_layout()

plt.show()
```

**Conclusion - The majority of customers who do not churn tend to have services like PhoneService, InternetService (particularly DSL), and OnlineSecurity enabled. For services like OnlineBackup, TechSupport, and StreamingTV, churn rates are noticeably higher when these services are not used or are unavailable**
